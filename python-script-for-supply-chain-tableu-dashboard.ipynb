```python
import pandas as pd
from googletrans import Translator


def translate_text(text, target_language='en'):
    translator = Translator()
    try:
        translation = translator.translate(text, dest=target_language)
        return translation.text
    except Exception as e:
        print(f"Translation error: {e}")
        return text  
```

```python
print(translate_text("Estados Unidos"))
```

```python
countries_EN=np.array([],dtype="object")

countries_EN = np.array([translate_text(value) for value in df['Order Country'].unique()], dtype="object")

countries_EN
```

```python
country_dict = dict(zip(df['Order Country'].unique(), countries_EN))

country_dict
```

```python
df['Order Country'] = df['Order Country'].replace(country_dict)
df['Order Country']
```
